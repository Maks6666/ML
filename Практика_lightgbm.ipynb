# Завдання 2
Розділіть дані на тестові та тренувальні(цільова ознака `SalePrice`)

```python
from sklearn.model_selection import train_test_split

x = df.drop("SalePrice", axis="columns")
y = df["SalePrice"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Завдання 3
Створіть препроцесор для заповнення пропусків та натренуйте його

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder

num_col = x.select_dtypes(include = "number").columns
cat_col = x.select_dtypes(include = "object").columns


num_transformer = Pipeline(steps=[("imputing", SimpleImputer(strategy="most_frequent"))])

cat_transform = Pipeline(steps=[
    ("imputing", SimpleImputer(strategy="most_frequent")),
    ("encoding", OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

preprocessor = ColumnTransformer(transformers=[
                                  ("numerical", num_transformer, num_col),
                                  ("categorical", cat_transform, cat_col),
                              ])

preprocessor.set_output(transform='pandas')
preprocessor
```

```python
preprocessor.set_output(transform='pandas')


x_train_preprocessed = preprocessor.fit_transform(x_train)


x_test_preprocessed = preprocessor.transform(x_test)
```

# Завдання 4
Створіть модель та натренуйте її

```python
import lightgbm as lgb


model = lgb.LGBMRegressor(objective='mse',
                          n_estimators = 900,
                          learning_rate = 0.09)


```

# Завдання 5
Виведіть метрики, важливість онак та графік метрик на тренувальних та тестових даних

```python
model.fit(x_train_preprocessed, y_train,
          eval_set = [(x_test_preprocessed, y_test), (x_train_preprocessed, y_train)],
          eval_metric="mae")
```

# Завдання 6
Об'єднайте препроцесор та натреновану модель і збережіть її

```python
model = Pipeline([("preprocessor", preprocessor), ("model", model)])
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn import metrics
r2 = metrics.r2_score(y_test, y_pred)
r2
```

```python
def get_metrics(y_true, y_pred, name="model"):
  df = pd.DataFrame()

  df.loc["MAE", name] = metrics.mean_absolute_error(y_true, y_pred)
  df.loc["RMSE", name] = metrics.mean_squared_error(y_true, y_pred) ** 0.5
  df.loc["R2", name] = metrics.r2_score(y_true, y_pred)

  return df.round(2)
```

```python
all_metrics = pd.DataFrame()

all_metrics["train"] = get_metrics(y_train, model.predict(x_train))
all_metrics["test"] = get_metrics(y_test, model.predict(x_test))

all_metrics
```
