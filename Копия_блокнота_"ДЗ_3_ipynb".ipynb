# Завдання 6
Перевірте чи правильно розраховано загальну вартість покупки. Виведіть ті рядки де є помилка. Виправте дані в таблиці

```python
def check_df(df):
    correct = True
    for index, row in df.iterrows():
        if row['units'] * row['unit_price'] != row['sale_amt']:
            print(f"Units: {row['units']}, Unit Price: {row['unit_price']}, Sale Amount: {row['sale_amt']}")
            correct = False
    if correct:
        print("DF is correct")

check_df(df)
```
